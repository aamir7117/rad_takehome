```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_json('data/data_analysis.json')
```

```python
df.head(200)
```

# EDA

```python
irrelevant_entries = ['None', 'none', 'null', '', ' ','0',0]
print "\nCounts of NoneType entries per field. Opposite of Fill rate\n", df.isnull().sum()
print "\nCounts of irrelevant entries per field\n", df.isin(irrelevant_entries).sum()

print "\nSetting all above entries to -1 so they're identifiable later"
df.fillna(-1,inplace=True)
df[df.isin(irrelevant_entries)] = -1
print "Done! All null and irrelevant entries updated to -1"
```

## Downloaded Zipcodes' Latitudes and Longitudes
* https://www.aggdata.com/node/86

```python
zipcodes = pd.read_csv('data/us_postal_codes.csv')
```

```python
df.zip = pd.to_numeric(df.zip)
big_df = df.join(zipcodes.set_index('Zip Code'),how='inner',on='zip')  
# Lost 120 records that didn't match real zipcodes. Still have close to 1 million records
```

## Filled 209 missing values by cross referencing zipcodes dataset
* 105 entries in the City field
* 104 entries in the State field

```python
df['city'][df.city<=0] = big_df['Place Name'].loc[df['city'][df.city<=0].index]
df['state'][df.state<=0] = big_df['State.1'].loc[df['state'][df.state<=0].index]
```

## Spatial distribution of businesses, observe US outline with specs in Hawaii and Alaska

```python
plt.figure(figsize=(16,9)); plt.xlabel('Longitude'),plt.ylabel('Latitude'); plt.title('Geo-locations of all the businesses; US outline clearly observed')
_ = plt.scatter(big_df['Longitude'],big_df['Latitude'],c='g')
```

## Revenue tiers shown over spatial location on mainland US
* Radius clients interested in B2B clients of a particular revenue range can reference this plot to find dense areas of potential clients

```python
data = big_df[big_df.revenue>0][['Latitude','Longitude','revenue']]
rev_order = ['Less Than $500,000', '$500,000 to $1 Million', '$1 to 2.5 Million',
             '$2.5 to 5 Million', '$5 to 10 Million', '$10 to 20 Million', '$20 to 50 Million',
             '$50 to 100 Million', '$100 to 500 Million', 'Over $500 Million', 'Over $1 Billion']
fg = sns.FacetGrid(data=big_df[['Latitude','Longitude','revenue']], hue='revenue', hue_order=rev_order,size=8, aspect=1.61,xlim=(-130,-60),ylim=(20,55))
_ = fg.map(plt.scatter, 'Longitude', 'Latitude').add_legend()
```

# Story - revenue levels and location  (see breakout plots below)

* As one moves from the East coast to the West coast, average revenue levels subside dramatically because most high revenue businesses are situated around the East coast. California is an exception
* This makes sense because the cities on the East coast are much older historically and have had more time to develop economically. 
* This means more time_in_business will correlate with higher revenues. We will check that next

```python
fg_1 = sns.FacetGrid(data=data, row='revenue',row_order=rev_order, hue='revenue',size=8, aspect=1.61,xlim=(-130,-60),ylim=(20,55))
_ = fg_1.map(plt.scatter, 'Longitude', 'Latitude').add_legend()
```
